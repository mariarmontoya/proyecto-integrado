#CLASS

**Class Usuario:**
Esta clase almacena la información de los usuarios registrados dentro del sistema, como nombre, identificación y datos personales.

```python
class clsUsuarios:
    def __init__(self, nombre, apellido, documento, correo):
        self.nombre = nombre
        self.apellido = apellido
        self.documento = documento
        self.correo = correo
```

**Class Item:**
Esta clase representa los elementos disponibles para préstamo y guarda información relacionada con cada item.

```python
class clsItem:
  def __init__(self, nombre,categoria, precio, id_item, estado):
    self.nombre = nombre
    self.categoria = categoria
    self.precio = precio
    self.id_item = id_item
    self.disponible = True

    def mostrar_item(self):
      print("ID:", self.id_item)
      print("Nombre:", self.nombre)
      print("categoria:", self.categoria)
      print("Precio:", self.precio)
      print("Estado:", self.estado)
```

**Class Prestamo:**
Esta clase registra la información de los préstamos realizados por los usuarios del sistema.

```python
class clsPrestamo:

    def __init__(self, usuario, item, fecha_prestamo):
        self.usuario = usuario
        self.item = item
        self.fecha_prestamo = fecha_prestamo
        self.fecha_devolucion = None
        self.activo = True

    def registrar_devolucion(self, fecha_devolucion):
        self.fecha_devolucion = fecha_devolucion
        self.activo = False
        self.item.disponible = True

    def calcular_dias(self):

        from datetime import datetime

        fecha_inicio = datetime.strptime(self.fecha_prestamo, "%d/%m/%Y")
        fecha_fin = datetime.strptime(self.fecha_devolucion, "%d/%m/%Y")

        diferencia = fecha_fin - fecha_inicio

        return diferencia.days

    def mostrar_prestamo(self):
        print("Usuario:", self.usuario.nombre)
        print("Item:", self.item.nombre)
```

**Class Devolucion:**
Esta clase controla y registra el proceso de devolución de los elementos prestados.

```python
class clsDevolucion:
  def __init__(self, usuario, item, fecha_prestamo, fecha_devolucion,
               fecha_actual, estado):

    self.usuario = usuario
    self.item = item
    self.fecha_prestamo = fecha_prestamo
    self.fecha_devolucion = fecha_devolucion
    self.fecha_actual = fecha_actual
    self.estado = estado

    def registrar_devolucion(self):
      self.item.disponible = True
      print("Devolucion registrada correctamente")

    def mostrar_devolucion(self):
        print("Usuario:", self.usuario.nombre)
        print("Item:", self.item.nombre)
        print("Fecha prestamo:", self.fecha_prestamo)
        print("Fecha devolucion:", self.fecha_devolucion)
        print("Fecha actual:", self.fecha_actual)
        print("Estado del producto:", self.estado)
```

#VALIDACIONES

**Función validar_texto:**
Esta función verifica que el texto ingresado sea válido y no contenga errores.

```python
def validar_texto(mensaje):

    while True:

        texto = input(mensaje).strip()

        if texto.replace(" ", "").isalpha():
            return texto.title()

        else:
            print("ERROR: Ingrese solo letras")
```

**Función validar_numero:**
Esta función comprueba que el dato ingresado sea numérico.

```python
def validar_numero(mensaje):

    while True:

        dato = input(mensaje).strip()

        if dato.isdigit():
            return int(dato)

        else:
            print("ERROR: Ingrese solo números")
```

**Función validar_correo:**
Esta función valida que el correo electrónico ingresado sea correcto.

```python
def validar_correo():

    while True:

        correo = input("Correo: ").strip()

        if "@" in correo and "." in correo:
            return correo

        else:
            print("ERROR: Ingrese un correo válido")
```

**Función validar_fecha:**
Esta función verifica que la fecha ingresada tenga un formato válido.

```python
def validar_fecha(mensaje):

    while True:

        fecha = input(mensaje)

        try:

            datetime.strptime(fecha, "%d/%m/%Y")

            return fecha

        except:
            print("ERROR: Fecha inválida. Use dd/mm/yyyy")
```

#FUNCIONES PRINCIPALES

**Función registrar_usuario:** Esta función permite agregar nuevos usuarios al sistema almacenando sus datos principales.

```python
def registrar_usuario():

  while True:

    nombre = input("Nombre:")
    if len(nombre) >= 3 and nombre.isalpha():
      break
    else:
     print("Error: El nombre debe tener al menos 3 caracteres y solo contener letras.")

  while True:

    apellido = input("Apellido:")
    if len(apellido) >= 3 and apellido.isalpha():
      break
    else:
      print("Error: El apellido debe tener minimo 3 letras y no numeros")

  while True:
    documento = input("Documento:")
    if documento.isdigit() and 3 <= len(documento) <= 15:
      break
    else:
      print("Error: El documento debe tener entre 3 y 15 digitos")

  while True:
    correo = input("Correo:")
    if "@" in correo and correo.endswith(".com"):
      break
    else:
      print("Error: Correo invalido")

  nuevo_usuario = clsUsuarios(nombre, apellido, documento, correo)

  usuario.append(nuevo_usuario)

  print("Usuario registrado correctamente")
```

**Función registrar_item:** Esta función se encarga de registrar los elementos disponibles para préstamo dentro del inventario.

```python
def registrar_item():

  while True:
    nombre = input("Nombre del item:")
    if len(nombre) >= 3:
      break
    else:
      print("Error: minimo 3 caracteres")

  categorias_disponibles = ["Videojuegos", "Libros", "Musica y video",
                            "Herramientas", "Dinero", "Miscelaneo"]
  categoria_seleccionada = ""

  while True:
    print("\nSeleccione la categoria:")
    for i, c in enumerate(categorias_disponibles, start=1):
      print(f"{i} - {c}")

    try:
      op = int(input("Ingrese el numero de la categoria:"))
      if 1 <= op <= len(categorias_disponibles):
        categoria_seleccionada = categorias_disponibles[op - 1]
        break
      else:
        print("Error: Opcion de categoria invalida. Intente de nuevo.")
    except ValueError:
      print("Error: Por favor, ingrese un número.")

  while True:
    try:
      precio_str = input("Precio: ").replace(".", "").replace(",", "").strip()

      while not precio_str.isdigit():

          print("ERROR: Ingrese solo números")

          precio_str = input("Precio: ").replace(".", "").replace(",", "").strip()

      precio = int(precio_str)
      if precio > 0:
        break
      else:
        print("Error: El precio debe ser un número positivo.")
    except ValueError:
      print("Error: Por favor, ingrese un número válido para el precio.")

  while True:
    id_item = input("ID del item:")
    if len(id_item) >= 3:
      break
    else:
      print("Error: El ID del item debe tener al menos 3 caracteres.")

  estado = input("Estado del item (Excelente, Bueno, Inaceptable):")

  item = clsItem(nombre, categoria_seleccionada, precio, id_item, estado)

  items.append(item)

  print("Item registrado correctamente")
```

**Función realizar_prestamo:** Esta función verifica la disponibilidad del item y registra el préstamo realizado por el usuario.

```python
def registrar_prestamo():

    doc = input("Documento del usuario: ")

    usuario_encontrado = None

    for u in usuario:
        if u.documento == doc:
            usuario_encontrado = u
            break

    if usuario_encontrado is None:
        print("Usuario no encontrado")
        return

    while True:

        id_item = input("ID del item: ")

        item_encontrado = None

        for i in items:
            if i.id_item == id_item:
                item_encontrado = i
                break

        if item_encontrado is not None:
            break

        print("Item no encontrado. Intente nuevamente.")

    if item_encontrado.disponible == False:
        print("El item ya está prestado")
        return

    while True:

        fecha_prestamo = input("Ingrese la fecha del préstamo (dd/mm/yyyy): ")

        try:
            datetime.strptime(fecha_prestamo, "%d/%m/%Y")
            break

        except ValueError:
            print("Fecha inválida")

    nuevo_prestamo = clsPrestamo(usuario_encontrado,item_encontrado,
                                 fecha_prestamo)
    prestamos.append(nuevo_prestamo)
    historial_prestamos.append(nuevo_prestamo)

    item_encontrado.disponible = False

    print("Préstamo registrado correctamente")
```

**Función devolver_item:** Esta función actualiza el estado del item cuando es devuelto y finaliza el préstamo.

```python
def registrar_devolucion():

    id_item = input("Ingrese el ID del item a devolver: ")

    prestamo_encontrado = None

    for p in prestamos:

        if p.item.id_item == id_item and p.activo:
            prestamo_encontrado = p
            break

    if prestamo_encontrado is None:
        print("No existe préstamo activo")
        return

    while True:

        fecha_devolucion = input("Ingrese la fecha de devolución (dd/mm/yyyy):")

        try:
            fecha_fin = datetime.strptime(fecha_devolucion, "%d/%m/%Y")
            fecha_inicio = datetime.strptime(prestamo_encontrado.fecha_prestamo,
                                             "%d/%m/%Y")

            if fecha_fin < fecha_inicio:
                print("La fecha de devolución no puede ser menor a la fecha del préstamo")
            else:
                break

        except ValueError:
            print("Fecha inválida")

    prestamo_encontrado.registrar_devolucion(fecha_devolucion)

    dias_prestado = prestamo_encontrado.calcular_dias()

    # =========================
    # ESTADO DEL ARTÍCULO
    # =========================

    while True:

        print("\nEstado del artículo:")
        print("1. Excelente")
        print("2. Bueno")
        print("3. Malo")

        estado = input("Seleccione una opción: ")

        if estado in ["1", "2", "3"]:
            break

        print("Valor inválido. Debe ingresar 1, 2 o 3.")

    # =========================
    # VENTA POR RETRASO Y/O MAL ESTADO
    # =========================

    precio_base = prestamo_encontrado.item.precio

    recargo_porcentaje = 0
    motivos = []

    if dias_prestado > 30:

        recargo_porcentaje += 0.23
        motivos.append("Más de 30 días")

    if estado == "3":

        recargo_porcentaje += 0.23
        motivos.append("Mal estado")

    if recargo_porcentaje > 0:

        recargo_total = precio_base * recargo_porcentaje
        total = precio_base + recargo_total

        venta = {"item": prestamo_encontrado.item.nombre,"precio": precio_base,
                 "recargo": recargo_total,"total": total,"motivo":
                 " + ".join(motivos)}

        ventas.append(venta)

        print("\n===== VENTA GENERADA =====")
        print(f"Artículo: {prestamo_encontrado.item.nombre}")
        print(f"Motivo: {' + '.join(motivos)}")
        print(f"Precio base: ${precio_base:,.0f}")
        print(f"Recargo: ${recargo_total:,.0f}")
        print(f"Total: ${total:,.0f}")

        generar_factura(prestamo_encontrado)

        prestamos.remove(prestamo_encontrado)
        prestamo_encontrado.item.disponible = True

        return

    # =========================
    # DEVOLUCIÓN NORMAL
    # =========================

    prestamos.remove(prestamo_encontrado)
    prestamo_encontrado.item.disponible = True

    generar_certificado(prestamo_encontrado)

    print("\nDevolución registrada correctamente")
```

**Función consultar_prestamos:** Esta función muestra en pantalla todos los préstamos registrados en el sistema.

```python
def consultar_prestamos():

    print("\n=== LISTA DE PRESTAMOS ===")

    if len(historial_prestamos) == 0:
        print("No hay préstamos")
        return

    for p in historial_prestamos:

        print(f"\nUsuario: {p.usuario.nombre}")

        print(f"Item: {p.item.nombre}")

        print(f"Fecha préstamo: {p.fecha_prestamo}")

        if p.fecha_devolucion is not None:

            print(f"Fecha devolución: {p.fecha_devolucion}")

            fecha_inicio = datetime.strptime(p.fecha_prestamo,"%d/%m/%Y")

            fecha_fin = datetime.strptime(p.fecha_devolucion,"%d/%m/%Y")

            dias = (fecha_fin - fecha_inicio).days

            print(f"Días prestado: {dias}")

        else:

            print("El item aún no ha sido devuelto")
```

FUNCIONES DE REPORTE

**Función generar_pdf:**
Esta función crea un reporte en PDF con la información almacenada del sistema.

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet

def generar_factura(prestamo):

    try:

        precio = prestamo.item.precio

        impuesto = precio * 0.23

        total = precio + impuesto

        nombre_archivo = f"factura_{prestamo.usuario.documento}.pdf"

        pdf = SimpleDocTemplate(nombre_archivo)

        estilos = getSampleStyleSheet()

        contenido = []

        contenido.append(Paragraph("FACTURA DE VENTA", estilos["Title"]))
        contenido.append(Spacer(1,12))

        contenido.append(Paragraph(f"Cliente: {prestamo.usuario.nombre}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"Documento: {prestamo.usuario.documento}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"Artículo: {prestamo.item.nombre}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"Fecha préstamo: {prestamo.fecha_prestamo}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"Fecha devolución: {prestamo.fecha_devolucion}",
                                   estilos["Normal"]))

        contenido.append(Spacer(1,12))

        contenido.append(Paragraph("Motivo: préstamo superior a 30 días",
                                   estilos["Normal"]))

        contenido.append(Spacer(1,12))

        contenido.append(Paragraph(f"Subtotal: ${precio:,.0f}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"Impuesto (23%): ${impuesto:,.0f}",
                                   estilos["Normal"]))
        contenido.append(Paragraph(f"TOTAL: ${total:,.0f}",
                                   estilos["Normal"]))

        pdf.build(contenido)

        print(f"\nPDF generado correctamente: {nombre_archivo}")

    except Exception as e:

        print("Error generando PDF:")
        print(e)
```

**Función generar_certificado:**
Esta función genera certificados relacionados con usuarios o préstamos realizados.

```python
def generar_certificado(prestamo):

  from datetime import datetime

  nombre_archivo = prestamo.usuario.documento + "_" + prestamo.item.id_item + ".txt"

  fecha= datetime.now().strftime("%d/%m/%Y")

  with open(nombre_archivo, "w") as archivo:

    archivo.write("==== CERTIFICADO DE DEVOLUCION ====\n\n")
    archivo.write("Usuario: " + prestamo.usuario.nombre + "\n")
    archivo.write("Documento: " + prestamo.usuario.documento + "\n")
    archivo.write("Item: " + prestamo.item.nombre + "\n")
    archivo.write("Fecha prestamo: " + prestamo.fecha_prestamo + "\n")
    archivo.write("Fecha devolucion: " + fecha + "\n")

    print("Certificado generado:", nombre_archivo)

    from datetime import datetime
    fecha = datetime.now().strftime("%d/%m/%Y")
```

**Función exportar_csv:**
Esta función exporta la información almacenada en archivos CSV para facilitar su manejo.

```python
def exportar_csv():

    with open("prestamos.csv", "w", newline="", encoding="utf-8") as archivo:

        writer = csv.writer(archivo)

        writer.writerow(["Usuario","Documento","Item","Fecha Prestamo",
                         "Fecha Devolucion"])

        for p in prestamos:

            writer.writerow([p.usuario.nombre,p.usuario.documento,p.item.nombre,
                             p.fecha_prestamo,p.fecha_devolucion])

    print("CSV generado correctamente")
    print("Archivo guardado como: prestamos.csv")
```

**Función generar_reporte_admin:**
Esta función crea un reporte general con la información administrativa del sistema.

```python
def generar_reporte_admin():

    pdf = SimpleDocTemplate("reporte_administrador.pdf")
    estilos = getSampleStyleSheet()
    contenido = []

    # =========================
    # ESTADÍSTICAS GENERALES
    # =========================

    cantidad_ventas = len(ventas)

    total_ventas = 0
    for venta in ventas:
        total_ventas += venta.get("total", 0)

    contenido.append(Paragraph("REPORTE GENERAL DEL SISTEMA", estilos["Title"]))
    contenido.append(Spacer(1,12))

    contenido.append(Paragraph(f"Usuarios registrados: {len(usuario)}",
                               estilos["Normal"]))
    contenido.append(Paragraph(f"Items registrados: {len(items)}",
                               estilos["Normal"]))
    contenido.append(Paragraph(f"Préstamos registrados: {len(historial_prestamos)}",
                               estilos["Normal"]))
    contenido.append(Paragraph(f"Ventas realizadas: {cantidad_ventas}",
                               estilos["Normal"]))
    contenido.append(Paragraph(f"Valor total vendido: ${total_ventas:,.0f}",
                               estilos["Normal"]))

    contenido.append(Spacer(1,20))

    # =========================
    # USUARIOS REGISTRADOS
    # =========================

    contenido.append(Paragraph("TODOS LOS USUARIOS REGISTRADOS",
                               estilos["Heading1"]))

    for u in usuario:
        contenido.append(Paragraph(f"{u.nombre} {u.apellido} - Documento: {u.documento}",
                                   estilos["Normal"]))

    contenido.append(Spacer(1,20))

    # =========================
    # ITEMS REGISTRADOS
    # =========================

    contenido.append(Paragraph("TODOS LOS ITEMS REGISTRADOS",
                               estilos["Heading1"]))

    for item in items:
      contenido.append(Paragraph(f"{item.nombre} - Precio: ${item.precio:,.0f}",
                                 estilos["Normal"]))

    contenido.append(Spacer(1,20))

    # =========================
    # PRESTAMOS POR USUARIO
    # =========================

    conteo = {}

    # Inicializar todos los usuarios en 0
    for u in usuario:
        conteo[u.documento] = {"nombre": f"{u.nombre} {u.apellido}","cantidad": 0}

    for p in prestamos:
        conteo[p.usuario.documento]["cantidad"] += 1

    ranking = sorted(conteo.items(),key=lambda x: x[1]["cantidad"],reverse=True)

    contenido.append(Paragraph("USUARIOS QUE MÁS PRESTARON", estilos["Heading1"]))

    for p in historial_prestamos:conteo[p.usuario.documento]["cantidad"] += 1

    for doc, datos in ranking:
        contenido.append(Paragraph(f"{datos['nombre']} - {datos['cantidad']} préstamos",
                                   estilos["Normal"]))

    contenido.append(Spacer(1,20))

    contenido.append(Paragraph("USUARIOS QUE MENOS PRESTARON",
                               estilos["Heading1"]))

    ranking_menor = sorted(conteo.items(),key=lambda x: x[1]["cantidad"])

    for doc, datos in ranking_menor:
       contenido.append(Paragraph(f"{datos['nombre']} - {datos['cantidad']} préstamos",
                                  estilos["Normal"]))

    contenido.append(Spacer(1,20))

    # =========================
    # ITEMS PRESTADOS Y NO PRESTADOS
    # =========================

    contenido.append(Paragraph("ESTADO DE LOS ITEMS", estilos["Heading1"]))

    items_prestados = set()

    for p in historial_prestamos:items_prestados.add(p.item.nombre)

    for item in items:

        estado = ("PRESTADO ALGUNA VEZ" if item.nombre in items_prestados else "NUNCA PRESTADO")

        contenido.append(Paragraph(f"{item.nombre} - {estado}",estilos["Normal"]))

    # =========================
    # ARTICULOS VENDIDOS
    # =========================

    contenido.append(Spacer(1,20))

    contenido.append(Paragraph("ARTÍCULOS VENDIDOS", estilos["Heading1"]))

    if len(ventas) == 0:

        contenido.append(Paragraph("No hay artículos vendidos.",
                                   estilos["Normal"]))

    else:

        for venta in ventas:

            motivo = venta.get("motivo", "No especificado")

            contenido.append(Paragraph(f"Artículo: {venta['item']} | "f"Motivo: {motivo} | "f"Total: ${venta['total']:,.0f}",estilos["Normal"]))

    pdf.build(contenido)

    print("Reporte PDF generado correctamente")

    from google.colab import files
    files.download("reporte_administrador.pdf")
```

**Función estadistica_usuarios:**
Esta función genera estadísticas relacionadas con los usuarios registrados y sus préstamos.

```python
def estadisticas_usuarios():

    conteo = {}

    for p in historial_prestamos:

        doc = p.usuario.documento

        if doc not in conteo:

          conteo[doc] = {"nombre": p.usuario.nombre,"cantidad": 0}

        conteo[doc]["cantidad"] += 1

    if len(conteo) == 0:

        print("No hay préstamos registrados")
        return

    ranking = sorted(conteo.items(),key=lambda x: x[1]["cantidad"],reverse=True)

    print("\n===== TOP USUARIOS =====")

    for doc, datos in ranking:

        print(f"{datos['nombre']} - {doc} - {datos['cantidad']} préstamos")
```

**Función administrador:**
Esta función permite acceder a las opciones administrativas y gestionar el sistema.

```python
def administrador():

    user = input("Usuario administrador: ")
    password = input("Contraseña: ")

    if user != "admin" or password != "1234":

        print("Acceso denegado")
        return

    print("\n===== PANEL ADMINISTRADOR =====")

    print(f"Usuarios registrados: {len(usuario)}")
    print(f"Items registrados: {len(items)}")
    print(f"Préstamos registrados: {len(historial_prestamos)}")

    # ESTADÍSTICAS DE VENTAS

    cantidad_ventas = len(ventas)

    total_ventas = 0

    for venta in ventas:

        if isinstance(venta, dict):

            if "total" in venta:
                total_ventas += venta["total"]

            elif "precio" in venta:
                total_ventas += venta["precio"]

    print(f"Ventas realizadas: {cantidad_ventas}")
    print(f"Valor total vendido: ${total_ventas:,.0f}")

    print("\n--- ARTÍCULOS VENDIDOS ---")

    if len(ventas) == 0:

        print("No hay artículos vendidos.")

    else:

        for venta in ventas:

            motivo = venta.get("motivo", "No especificado")

            print(f"Artículo: {venta['item']} | "f"Motivo: {motivo} | "f"Total: ${venta['total']:,.0f}")


    print("\n--- USUARIOS CON MÁS PRÉSTAMOS ---")

    estadisticas_usuarios()

    print("\nGenerando CSV...")

    exportar_csv()

    print("CSV generado correctamente")

    print("\nGenerando PDF administrador...")

    generar_reporte_admin()
```

**Función menu_principal:**
Esta función controla la navegación del programa mostrando las opciones disponibles al usuario.

```python
def menu():

  while True:

    print("\n==== MJ LEDGER ====")
    print("1. Registrar Usuario")
    print("2. Registrar Item")
    print("3. Registrar Prestamo")
    print("4. Registrar Devolucion")
    print("5. Consultar prestamos")
    print("6. Administrador")
    print("7. Salir")

    opcion = input("Seleccione opcion:")

    if opcion == "1":
      registrar_usuario()

    elif opcion == "2":
      registrar_item()

    elif opcion == "3":
      registrar_prestamo()

    elif opcion == "4":
      registrar_devolucion()

    elif opcion == "5":
      consultar_prestamos()

    elif opcion == "6":
      administrador()

    elif opcion == "7":
      print("Saliendo del sistema")
      break
```
